# handeling missing value

```python
df['last review']= pd.to_datetime(df['last review'], errors='coerce')
```

```python
df.info()
```

```python
df.fillna(
    {
        'reviews per month': 0,
        'last review': df['last review'].min()
    },
    inplace=True
)
```

```python
df.dropna(subset=['NAME','host name'],inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df=df.drop(columns=["license","house_rules"],errors='ignore')
```

```python
df.head()
```

# remove doller sign from price and service fee

```python
df['price'] = df['price'].replace(r'[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace(r'[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

# remove duplicates

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

# visualization 
 what is the distribution of listing price


```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'],bins=50,kde=True,color='red')
plt.title("Distribution of listing price")
plt.xlabel("price $")
plt.ylabel("Frequency")
plt.show()
```

The histograms shows a fairly even distribution of listing prices across different price ranges,indicating no particular concentration of listing in any specific price range.the KDE line helps visualize this even spread more clearly
confirming that the datasetcontains listing with a wide variety of prices.

# how are different room types distributed

```python
df['room type']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="room type",data=df,color='hotpink')
plt.title("room type distribution")
plt.xlabel("room type")
plt.ylabel("count")
plt.show()
```

# how are listing distribution across different neighbours

```python
plt.figure(figsize=(12,8))
sns.countplot(
    y='neighbourhood group',
    data=df,
    color="lightgreen",
    order=df['neighbourhood group'].value_counts().index
)
plt.title('Number of listings by neighbourhood group')
plt.xlabel('Count')
plt.ylabel('Neighbourhood Group')
plt.show()
```

# what is the relationship between price and room type

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='room type',y='price',hue='room type',data=df,palette='Set1')
plt.title('price vs room type')
plt.xlabel('room type')
plt.ylabel('price($)')
plt.legend(title='room type')
plt.show()
```

# how has the number of reviews changed over time

```python
df['last review']=pd.to_datetime(df['last review'])
reviews_over_time =df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize=(12,6))
reviews_over_time.plot(kind='line',color='red')
plt.title('Number of reviews over time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
